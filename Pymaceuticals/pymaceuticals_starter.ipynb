```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load
mouse_drug_data = "data/mouse_drug_data.csv"
clinical_trial_data = "data/clinicaltrial_data.csv"
```

```python
# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug=pd.read_csv(mouse_drug_data)
mouse_drug.head()
```

```python
clinical_trial=pd.read_csv(clinical_trial_data)
clinical_trial.head()
```

```python
# Combine the data into a single dataset
combined_data=pd.merge(clinical_trial,mouse_drug,on="Mouse ID",how='outer')

# Display the data table for preview
combined_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_df=combined_data.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].mean()
# Convert to DataFrame
mean_df=mean_df.reset_index()
# Preview DataFrame
mean_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
se_df=combined_data.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].sem()

# Convert to DataFrame
se_df=se_df.reset_index()
# Preview DataFrame

se_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_pivot=mean_df.pivot(index='Timepoint',columns='Drug',values='Tumor Volume (mm3)')

# Preview that Reformatting worked
mean_pivot
```

```python
se_pivot=se_df.pivot(index='Timepoint',columns='Drug',values='Tumor Volume (mm3)')
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(mean_pivot.index, mean_pivot['Capomulin'], se_pivot['Capomulin'], 
             fmt='o', linestyle='dotted', label='Capomulin', color='r')
plt.errorbar(mean_pivot.index, mean_pivot['Infubinol'], se_pivot['Infubinol'], 
             fmt='^', linestyle='dotted', label='Infubinol', color='b')
plt.errorbar(mean_pivot.index, mean_pivot['Ketapril'], se_pivot['Ketapril'], 
             fmt='s', linestyle='dotted', label='Ketapril', color='g')
plt.errorbar(mean_pivot.index, mean_pivot['Placebo'], se_pivot['Placebo'], 
             fmt='h', linestyle='dotted', label='Placebo', color='k')

plt.legend(loc='best')
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid()

# Save the Figure

plt.savefig("../Images/TumorResponse.png")
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_df2=combined_data.groupby(['Drug','Timepoint'])['Metastatic Sites'].mean()

# Convert to DataFrame
mean_df2=mean_df2.reset_index()
# Preview DataFrame
mean_df2.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
se_df2=combined_data.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].sem()
# Convert to DataFrame
se_df2=se_df2.reset_index()
# Preview DataFrame
se_df2.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_pivot2=mean_df2.pivot(index='Timepoint',columns='Drug',values='Metastatic Sites')

# Preview that Reformatting worked
mean_pivot2
```

```python
se_pivot2=se_df2.pivot(index='Timepoint',columns='Drug',values='Metastatic Sites')
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(mean_pivot2.index, mean_pivot2['Capomulin'], se_pivot2['Capomulin'], 
             fmt='o', linestyle='dotted', label='Capomulin', color='r')
plt.errorbar(mean_pivot2.index, mean_pivot2['Infubinol'], se_pivot2['Infubinol'], 
             fmt='^', linestyle='dotted', label='Infubinol', color='b')
plt.errorbar(mean_pivot2.index, mean_pivot2['Ketapril'], se_pivot2['Ketapril'], 
             fmt='s', linestyle='dotted', label='Ketapril', color='g')
plt.errorbar(mean_pivot2.index, mean_pivot2['Placebo'], se_pivot2['Placebo'], 
             fmt='h', linestyle='dotted', label='Placebo', color='k')


plt.legend(loc='best')
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment duration (Days)")
plt.ylabel("Met. Sites")
plt.grid()
# Save the Figure
plt.savefig("../Images/MetastaticSpread.png")
# Show the Figure
plt.show()
```
